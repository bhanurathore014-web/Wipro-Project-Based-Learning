```python
# write a program to check if a given number is positive negative or zero.

num = int(input("Enter num: "))
if num > 0:
    print("positive")
elif num<0:
    print("negative")
else:
    print("zero")
```

```python
# write a program to check if a given number is odd or even.

num = int(input("Enter num: "))
if num%2==0:
    print("even")
else:
    print("odd")
```

```python
# Given two non-negative values, print true if they same last digit such as 27 and 57.
# lastdigit(7,17) → true
# lastdigit(6,17) → false
# lastdigit(3,113) → true

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

if a % 10 == b % 10:
    print(True)
else:
    print(False)
```

```python
# write a program to check if a given number is prime or not.
n = int(input("Enter a number: "))
for i in range(2, n):
    if n%i==0:
        print("Not prime")
        break
else:
    print("Prime")
```

```python
# wap to print prime numbers between 10 to 99.
for i in range(10,100):
    for j in range(2,i):
        if i%j==0:
            break
    else:
        print(i, end=" ")
```

```python
# write a program to find the sum of digits of a number. 
# I/P = 1234
# O/P = 10

num=1234
sum=0
while num>0:
    sum+=num%10
    num//=10
print(sum)
```

```python
# wap to reverse a number.
# I/P = 1234
# O/P = 4321
num = int(input("Enter a number: "))
reverse = 0
while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num //= 10
print("Reversed number:", reverse)
```

```python
# wap to find if given number is palindrome or not.
# I/P = 11011
# O/P = Palindrome
num = int(input("Enter a number: "))
original_num = num
reverse = 0
while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num //= 10
if original_num == reverse:
    print("Palindrome")
else:
    print("Not Palindrome")
```
